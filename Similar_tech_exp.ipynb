```python
from nltk.corpus import stopwords
from nltk.cluster.util import cosine_distance
import numpy as np
import networkx as nx
 
def read_article(file_name):
    file = open(file_name, "r")
    filedata = file.readlines()
    article = filedata[0].split(". ")
    sentences = []

    for sentence in article:
        print(sentence)
        sentences.append(sentence.replace("[^a-zA-Z]", " ").split(" "))
    sentences.pop() 
    
    return sentences

def sentence_similarity(sent1, sent2, stopwords=None):
    if stopwords is None:
        stopwords = []
 
    sent1 = [w.lower() for w in sent1]
    sent2 = [w.lower() for w in sent2]
 
    all_words = list(set(sent1 + sent2))
 
    vector1 = [0] * len(all_words)
    vector2 = [0] * len(all_words)
 
    # build the vector for the first sentence
    for w in sent1:
        if w in stopwords:
            continue
        vector1[all_words.index(w)] += 1
 
    # build the vector for the second sentence
    for w in sent2:
        if w in stopwords:
            continue
        vector2[all_words.index(w)] += 1
 
    return 1 - cosine_distance(vector1, vector2)
 
def build_similarity_matrix(sentences, stop_words):
    # Create an empty similarity matrix
    similarity_matrix = np.zeros((len(sentences), len(sentences)))
 
    for idx1 in range(len(sentences)):
        for idx2 in range(len(sentences)):
            if idx1 == idx2: #ignore if both are same sentences
                continue 
            similarity_matrix[idx1][idx2] = sentence_similarity(sentences[idx1], sentences[idx2], stop_words)

    return similarity_matrix


def generate_summary(file_name, top_n=5):
    stop_words = stopwords.words('english')
    summarize_text = []

    # Step 1 - Read text anc split it
    sentences =  read_article(file_name)

    # Step 2 - Generate Similary Martix across sentences
    sentence_similarity_martix = build_similarity_matrix(sentences, stop_words)

    # Step 3 - Rank sentences in similarity martix
    sentence_similarity_graph = nx.from_numpy_array(sentence_similarity_martix)
    scores = nx.pagerank(sentence_similarity_graph)

    # Step 4 - Sort the rank and pick top sentences
    ranked_sentence = sorted(((scores[i],s) for i,s in enumerate(sentences)), reverse=True)    
    print("Indexes of top ranked_sentence order are ", ranked_sentence)    

    for i in range(top_n):
      try:
        summarize_text.append(" ".join(ranked_sentence[i][1]))
      except:
        pass

    # Step 5 - Offcourse, output the summarize texr
    print("Summarize Text: \n", ". ".join(summarize_text))

# # let's begin
# generate_summary( "msft.txt", 2)
```

```python
from nltk.stem import WordNetLemmatizer

def text_preprocess(ds: pd.Series) -> pd.Series:
    """
    Apply NLP Preprocessing Techniques to the reviews.
    """
    for m in range(len(ds)):
        
        main_words = re.sub('[^a-zA-Z]', ' ', ds[m])                                      # Retain only alphabets
        main_words = (main_words.lower()).split()
        main_words = [w for w in main_words if not w in set(stopwords.words('english'))]  # Remove stopwords
        
        lem = WordNetLemmatizer()
        main_words = [lem.lemmatize(w) for w in main_words if len(w) > 1]                 # Group different forms of the same word
        
        main_words = ' '.join(main_words)
        ds[m] = main_words

    return ds
```

```python
def clean_text(
    string: str, 
    punctuations=r'''!()-[]{};:'"\,<>./?@#$%^&*_~''',
    stop_words=['the', 'a', 'and', 'is', 'be', 'will']) -> str:
    """
    A method to clean text 
    """
    # Cleaning the urls
    string = re.sub(r'https?://\S+|www\.\S+', '', string)

    # Cleaning the html elements
    string = re.sub(r'<.*?>', '', string)

    # Removing the punctuations
    for x in string.lower(): 
        if x in punctuations: 
            string = string.replace(x, "") 

    # Converting the text to lower
    string = string.lower()

    # Removing stop words
    string = ' '.join([word for word in string.split() if word not in stop_words])

    # Cleaning the whitespaces
    string = re.sub(r'\s+', ' ', string).strip()

    return string
```

```python
def get_cosine_similarity(feature_vec_1, feature_vec_2):
    return cosine_similarity(feature_vec_1.reshape(1, -1), feature_vec_2.reshape(1, -1))[0][0]
```

```python
def check_tech(tech_main, tech_comp):
    tech_flag = ''
    if len(tech_main) > len(tech_comp):
        if tech_main.find(tech_comp) != -1:
            tech_flag = 'tech_sub_string'
        else:
            nltk_tokens = nltk.word_tokenize(tech_comp)
            bi_gram = list(nltk.bigrams(nltk_tokens))
            bi_gram_lst = [' '.join(list(x)) for x in bi_gram]
            lst = [x for x in bi_gram_lst if tech_main.find(x) != -1]

            if len(lst) > 0:
                tech_flag = 'tech_ngram'
            else:
                tech_flag = 'no_sim'
                # tech_comp_ls = tech_comp.split()
                # lst = [x for x in tech_comp_ls if tech_main.find(x) != -1]
                # if len(lst) > 0:
                #     tech_flag = 'tech_ngram'
                # else:
                #     tech_flag = 'no_sim'
    else:
        if tech_comp.find(tech_main) != -1:
            tech_flag = 'tech_sub_string'
        else:
            nltk_tokens = nltk.word_tokenize(tech_main)
            bi_gram = list(nltk.bigrams(nltk_tokens))
            bi_gram_lst = [' '.join(list(x)) for x in bi_gram]
            lst = [x for x in bi_gram_lst if tech_comp.find(x) != -1]

            if len(lst) > 0:
                tech_flag = 'tech_ngram'

            else:
                tech_flag = 'no_sim'
                # tech_main_ls = tech_main.split()
                # lst = [x for x in tech_main_ls if tech_comp.find(x) != -1]
                # if len(lst) > 0:
                #     tech_flag = 'tech_ngram'
                # else:
                #     tech_flag = 'no_sim'

    return tech_flag
```

```python
def summarize(tech_dict):
    tech_list = tech_dict.keys()
    tech_dict_new = {}
    for key_tech in tech_dict.keys():
        tech = key_tech
        desc = tech_dict[key_tech]

        tech = tech.lower().replace('®', '').replace('-', '').replace('™', '').replace('\xa0', ' ').replace('\n',
                                                                                                            ' ').strip(
            '\xa0').strip('\n').strip()
        doc_ls = desc.lower().replace('®', '').replace('-', '').replace('™', '').replace('\xa0', ' ').strip().split(
            '. ')
        doc_final = '. '.join([x for x in doc_ls if x.find(tech) != -1])

        doc_final = doc_final.split('\n')

        desc_final = '\n'.join([x for x in doc_final if x.find(tech) != -1]).replace('..', '.').replace(
            '...', '.').replace('....', '.')

        # for tech_dom in tech_list:
        #     tech_dom = tech_dom.lower().replace('®', '').replace('-', '').replace('™', '').replace('\xa0', ' ').replace(
        #         '\n', ' ').strip('\xa0').strip('\n').strip()
        #
        #     if tech_dom != tech:
        #         if desc_final.find(tech_dom) != -1 and len(tech_dom) > len(tech):
        #             desc_final = desc_final.split('. ')
        #             desc_final = '. '.join([x for x in desc_final if x.find(tech_dom) == -1])

        desc_sm = summarizer(desc_final, truncation=True)[0]['summary_text']

        tech_dict_new[key_tech] = desc_sm

    return tech_dict_new
```

```python
def encode_data(tech_dict):
    emb_dict = {}

    for val in tech_dict.keys():
        tech = val
        desc = tech_dict[val]

        desc_emb = model.encode(desc)

        emb_dict[tech] = desc_emb

    return emb_dict
```

```python
def get_similar_tech(emb_dict):
    tech_lst = emb_dict.keys()

    final_lst = []
    for val in tech_lst:
        emb_eval = emb_dict[val]

        tech_main = str(val).lower().strip().replace('®', '').replace('-', '').replace('™', '').replace('\xa0',
                                                                                                        ' ').replace(
            '\n', ' ').strip('\xa0').strip('\n').strip()

        temp_lst = []
        for inner_val in tech_lst:
            tech_comp = str(inner_val).lower().strip().replace('®', '').replace('-', '').replace('™',
                                                                                                 '').replace(
                '\xa0', ' ').replace('\n', ' ').strip('\xa0').strip('\n').strip()
            if inner_val != val:
                if tech_comp != tech_main:
                    idx_agn = inner_val[0]
                    emb_agn = emb_dict[inner_val]
                    sim = get_cosine_similarity(emb_eval, emb_agn)

                    tech_flag = check_tech(tech_main, tech_comp)

                    print("compare the result here: ", tech_comp, tech_main, ':--> ',  sim, '---', tech_flag)

                    if tech_flag == 'tech_sub_string' and sim > 0.88:
                        temp_lst.append(inner_val)
                    elif tech_flag == 'tech_ngram' and sim > 0.84:
                        temp_lst.append(inner_val)
                    elif (tech_flag == '' or tech_flag == 'no_sim') and sim > 0.80:
                        temp_lst.append(inner_val)
                else:
                    temp_lst.append(inner_val)
        final_lst.append({val: temp_lst})

    return final_lst
```

```python
def get_sim_tech(tech_list):
    try:
        tech_dict = {}
        df = pd.read_excel('sim_tech_data.xlsx', sheet_name='Sheet1')
        tech_final_lst = []
        for val in tech_list:
            desc = str(df[df['Technology'] == val]['Description'].tolist()[0])
            tech_dict[val] = desc
        tech_dict_sum = summarize(tech_dict)
        print('tech_dict_sum', tech_dict_sum)
        tech_dict_emb = encode_data(tech_dict_sum)
        final_lst = get_similar_tech(tech_dict_emb)
        print('final list: ------------> ', final_lst)
        # temp_lst = []
        sim_lst = []
        tech_list = []
        for val in final_lst:
          if list(val.keys())[0] in sim_lst or any(item in sim_lst for item in list(val.values())[0]):
            sim_lst.append(list(val.keys())[0])
            sim_lst.extend(list(val.values())[0])
          else:
            tech_list.append(list(val.keys())[0])
            sim_lst.append(list(val.keys())[0])
            sim_lst.extend(list(val.values())[0])
        tech_list = list(set(tech_list))
        data = {'Similar_Tech': tech_list}
        return data
    except Exception as exe:
        print("Exception in get_sim_tech", exe)
```
